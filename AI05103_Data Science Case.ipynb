# Create Machine Learning Model to Predict Termd and Question

```python
#11. Buat fitur baru (kolom baru) berupa "Age" dalam satuan tahun. Untuk mendapatkannya, ambil data pada kolom DOB dikurangi dengan 2022
#(5 points)

data['DOB_date'] = pd.to_datetime(data['DOB'])
data['Age'] = 2022 - data['DOB_date'].dt.year

# Menampilkan data
data[['DOB', 'DOB_date', 'Age']]
```

```python
#12. Lakukan MinMaxScaler dari fitur "Salary"
# (5 points)
from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScaler instance
scaler = MinMaxScaler()

# Fit and transform the 'Salary' column
data['Salary_minmax'] = scaler.fit_transform(data[['Salary']])

data[['Salary','Salary_minmax']]
```

```python
#13. Ubahlah semua tipe data object menjadi category
# 5 point
data.dtypes
```

```python
# Looping kolom dan ubah tipe objek menjadi kategori
for column in data.select_dtypes(include=['object']):
    data[column] = data[column].astype('category')

# Check the data types after conversion
print(data.dtypes)
```

```python
#14. Ubahlah "RecruitmentSource" dengan menggunakana Label Encoder
# 5 point

from sklearn.preprocessing import LabelEncoder

# Inisialisasi LabelEncoder
recruitment_encoder = LabelEncoder()

# Sesuaikan dan ubah kolom 'RecruitmentSource'
data['RecruitmentSource'] = recruitment_encoder.fit_transform(data['RecruitmentSource'])

# Menampilkan beberapa baris pertama kumpulan data dengan kolom 'RecruitmentSource' yang dikodekan
print(data[['Employee_Name', 'RecruitmentSource']].head())
```

```python
#15. Ubahlah "PerformanceScore" dengan menggunakana Label Encoder
# 5 point

# Inisialisasi LabelEncoder
label_encoder = LabelEncoder()

# Sesuaikan dan ubah kolom 'PerformanceScore'
data['PerformanceScore'] = label_encoder.fit_transform(data['PerformanceScore'])

# Menampilkan beberapa baris pertama kumpulan data dengan kolom 'PerformanceScore' yang dikodekan
print(data[['Employee_Name', 'PerformanceScore']].head())
```

```python
data.dtypes
```

```python
#16. Pisahkan dependent variable dan independent variable
# 5 point

#Dependent Variable (Target)
y = data['Termd']

print('Dependent Variable:')
y.head(5)
```

```python
#Independent Variable (Feature)
X = data
X.drop(['Termd'], axis=1)
X['DOB_Date']=pd.to_datetime(X['DOB'])
X['Age'] = 2022 - X['DOB_Date'].dt.year
Salary = X['Salary'].to_numpy()
salary = Salary.reshape(-1,1)
scaler = MinMaxScaler()
hasil_scale = scaler.fit_transform(salary)
X['Salary_minmax']=hasil_scale
X = X[['MarriedID', 'MaritalStatusID', 'GenderID', 'DeptID', 'PerfScoreID', 'PositionID',
       'Age', 'RecruitmentSource', 'PerformanceScore', 'Salary_minmax', 'Absences', 'SpecialProjectsCount',
       'EmpSatisfaction', 'EngagementSurvey']]
print('Independent Variable:')
X.head(5)
```

```python
#17. Bagi data menjadi Train-Test sebesar 80:20
# (5 points)

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_recall_curve, roc_curve, precision_score, recall_score

# Memisahkan data menjadi set Train-Test sebesar 80:20
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=35)

# Print hasil train dan test
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
#18. Lakukan pemodelan machine learning (metode klasifikasi bebas) untuk bisa memprediksi status "Termd". Lakukan evaluasi hasil dengan menampilkan
# score klasifikasi, precision dan recall.
# Gunakan model yang menurut kalian terbaik
# (5 points)

# Import LogisticRegression
from sklearn.linear_model import LogisticRegression

# Membuat objek model regresi logistik
model = LogisticRegression(tol=0.00001, C=1, max_iter=1000, penalty =None)

# Train model regresi logistik menggunakan data train
train_model = model.fit(X_train, y_train)

# Memprediksi nilai target
y_pred = train_model.predict(X_test)
predictions = [round(value) for value in y_pred]

# Menghitung akurasi, presisi, dan recall model
accuracy = accuracy_score(y_test, predictions)
precision = precision_score(y_test, predictions)
recall = recall_score(y_test, predictions)

# Print hasil dlm bentuk persen
print("Accuracy : %.2f%%" % (accuracy * 100.0))
print("Precision : %.2f%%" % (precision* 100.0))
print("Recall : %.2f%%" % (recall * 100.0))
```

Menurut saya, model terbaik yang diperoleh adalah model logistic regression yang mendapatkan akurasi sebesar 61.90%.

## 19.	Is there any relationship between who a person works for (their manager) and their performance score? (5 points)<br>

```python
# Load the data
dt = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/HRDataset_v14.csv')
dt.head()
```

```python
# Menghitung jumlah karyawan dalam setiap kategori Performance Score
dt['PerformanceScore'].value_counts()
```

```python
# Visualisasi hubungan antara manajer, performance score, dan jumlah karyawan dalam bentuk bar chart
b.plot(kind = 'bar')
plt.show()
```

Dari visualisasi di atas dapat dilihat bagaimana distribusi karyawan yang memiliki performance score lebih baik dibagi berdasarkan manajer mereka. Jika ada manajer tertentu yang memiliki jumlah karyawan dengan performance score yang lebih baik secara signifikan lebih banyak atau lebih sedikit dibandingkan dengan manajer lainnya, ini dapat menunjukkan adanya hubungan antara manajer dan performance score. Sebaliknya, jika distribusi ini relatif seragam di antara manajer, maka mungkin tidak ada hubungan yang kuat antara manajer dan performance score.

Jadi, dapat disimpulkan bahwa adanya hubungan antara manajer dan performance score karena manager dapat memberikan pengaruh yang baik kepada performa karyawan.
